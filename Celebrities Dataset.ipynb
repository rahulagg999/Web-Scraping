#### Importing Libraries

```python
from bs4 import BeautifulSoup as soup  # HTML data structure
import xlsxwriter
from io import BytesIO
from urllib.request import urlopen as uReq     # Web client
```

#### IMDB website url for fetching data

```python
page_url = "https://www.imdb.com/list/ls068010962"
```

```python
uClient = uReq(page_url)
```

```python
page_soup = soup(uClient.read(), "html.parser")
```

```python
containers = page_soup.findAll("div", {"class": "lister-item mode-detail"})
```

#### Create an new Excel file and add a worksheet.

```python
workbook = xlsxwriter.Workbook('Celebrities Info.xlsx')
worksheet = workbook.add_worksheet()
```

#### Customising cell size in worksheet

```python
worksheet.set_column('C:D', 25)
worksheet.set_default_row(160)
```

#### Creating Headers in worksheet

```python
row = 0
col = 0

worksheet.write(row, col,     'Name')
worksheet.write(row, col + 1, 'Designation')
worksheet.write(row, col + 2, 'Movie')
worksheet.write(row, col + 3, 'Image')
```

```python
row=0
col=0

import os

for container in containers:

    star_name = container.h3.a.text.strip()                 # Fetching Name of the Celebrities
    
    genre = container.p.text
    clear_genre = genre.split('|')[0].strip()               # Fetching Designation of Celebrities
    
    Best_Movie = container.p.text.split('|')[1].strip()     # Fetching name of the Movies
    
    row += 1
    all_images = container.find('img').attrs['src']         # Fetching images link from website  
    image_data = uReq(all_images).read()
    file = open("images/"+str(row)+"image.jpg", "wb")       # Downloading them into a folder
    file.write(image_data)
    file.close()
    
    worksheet.write(row, col,     star_name)
    worksheet.write(row, col + 1, clear_genre)              # Writing all the data into worksheet
    worksheet.write(row, col + 2, Best_Movie)
    worksheet.insert_image(row, col + 3, "images/"+str(row)+"image.jpg")
    
    print("Name : " + star_name + "\n")
    print("Genre : " + clear_genre + "\n")
    print("Best Movie : " + Best_Movie + "\n")
    print("Images : " + all_images + "\n")
    
workbook.close()    

# End of Code
```
